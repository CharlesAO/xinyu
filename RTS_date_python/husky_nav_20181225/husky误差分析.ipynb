```python
import pandas as pd
filename = 'date_result.csv'
import matplotlib.pylab as plt

outfile = 'out_plot.csv'

with open(filename, 'r') as f:
    rows = f.readlines()

df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df['sq_x_y'] = (df.x**2 + df.y**2)**0.5
```

```python
df_1 = df.shift(1)
df_1
```

```python
data = {'point':[1,2,3,4,5,6,7,8,9]}
```

```python
df_erro = pd.DataFrame(data)

df_erro['l_1'] = df_1.sq_x_y[0:9].reset_index(drop = True)
df_erro['l_2'] = df_1.sq_x_y[9:18].reset_index(drop = True)
df_erro['l_3'] = df_1.sq_x_y[18:27].reset_index(drop = True)
df_erro['l_4'] = df_1.sq_x_y[27:36].reset_index(drop = True)
df_erro['l_5'] = df_1.sq_x_y[36:].reset_index(drop = True)

df_erro['yaw_1'] = df_1.yaw[0:9].reset_index(drop = True)
df_erro['yaw_2'] = df_1.yaw[9:18].reset_index(drop = True)
df_erro['yaw_3'] = df_1.yaw[18:27].reset_index(drop = True)
df_erro['yaw_4'] = df_1.yaw[27:36].reset_index(drop = True)
df_erro['yaw_5'] = df_1.yaw[36:].reset_index(drop = True)
```

```python
df_erro
```

```python
plt.plot(df_erro.point, df_erro.l_1, kind = ‘scatter’)
```
